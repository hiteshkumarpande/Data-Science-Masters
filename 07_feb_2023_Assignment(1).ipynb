# Q-1) You are writing code for a company. The requirement of the company is that you create a python function that will check whether the password entered by the user is correct or not. The function should take the password as input and return the string "Valid Password" if the entered password follows the below-given password guidelines else it should return "invalid Password".

# Note:
#    1. The Password should contain at least two uppercase letters and at least twolowercase letters.
#    2. The Password should contain at least a number and three specail characters.
#    3. The length of the password should be 10 characters long.


```python
def check_password(password):
    if len(password)!=0:
        return "Invalid Password"
    else:
        
        uppercase_chr = 0
        lowercase_chr = 0
        specail_chr = 0
        number_cnt = 0

        for i in password:
            if i.isupper():
                uppercase_chr+=1
            elif i.islower():
                lowercase_chr+=1
            elif i.isdigit():
                number_cnt+=1
            else:
                specail_chr+=1


        if uupercase_chr>=2 and lowercase_chr>=2 and specail_chr>=3 and number_cnt>=1:
            return "Valid Password"
        else:
            return "invalid Password"
        
password = input('Enter your password:')
result = check_password(password)
print(result)
```

# a) Check if the string starts with a particular letter.

```python
starts_with_letter = lambda string, letter: string.startswith(letter)

# Example usage:
input_string = input("Enter the string:\n")
letter_to_check = input("Enter the letter to check:\n")

if starts_with_letter(input_string, letter_to_check):
    print(f"The string starts with '{letter_to_check}'.")
else:
    print(f"The string does not start with '{letter_to_check}'.")
```

# b) Check if the string is numeric.

```python
to_numeric = lambda string: float(string) if string.replace(".","",1).isdigit() else None

# testing the lambda function with map()

strings = ["123","3.14","abc"]
numeric_values = list(map(to_numeric,strings))
print(numeric_values)
```

```python
strings=["123","3.14","abc"]
numeric_strings=[i for i in strings if i.isdigit()]
print(numeric_strings)
```

# c) Sort a list of tuples having fruit names and their quantity.
# [("mango",99),("orange",80),("grapes",1000)]

```python
l_o_t=[("mango",99),("orange",80),("grapes",1000)]

sorted_l_o_t=sorted(l_o_t,key=lambda l_o_t:l_o_t[1])
print(sorted_l_o_t)
```

# d) Find the squares of numbers from 1 to 10.

```python
# lambda and map()
squared_number = list(map(lambda x: x**2, range(1,11)))
print(squared_number)
```

```python
# lambda and filter()
square_num = list(filter(lambda x:x, map(lambda x:x**2, range(1,11))))
print(square_num)                 
```

# e) Find the cube root of numbers from 1 to 10.

```python
# lambda and filter
cube_root_num = list(filter(lambda x:x,map(lambda x: x**(1/3),range(1,11))))
print(cube_root_num)
```

```python
# lambda and map
cube_root = list(map(lambda x:x**(1/3),range(1,11)))
print(cube_root)
```

# f) Check if a number is even

```python
# lambda and filter
l=[]
n=int(input())
for i in range(1,n+1):
    l.append(i)
list(filter(lambda x:x%2==0,l))
```

```python
# list comprehension
y=[i for i in range(1,801) if i%2==0]
y
```

# h)  Sort a list of integers into positive and negative integers lists.

# [1,2,3,4,5,6,-1,-2,-3,-4,-5,0]

```python
l=[1,2,3,4,5,6,-1,-2,-3,-4,-5,0]
l_positive=[]
l_negative=[]
for i in l:
    if i>=0:
        l_positive.append(i)
        
    else:
        l_negative.append(i)
print('List of positive integers:',l_positive)
print("List of negative integers:",l_negative)
```
