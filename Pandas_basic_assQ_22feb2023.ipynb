# Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data rows and columns.

## Differences:

```python
import pandas as pd

# Example DataFrame
data = {'Name': ['John', 'Anna', 'Peter', 'Linda'],
        'Age': [25, 35, 30, 28],
        'City': ['New York', 'Paris', 'London', 'Sydney']}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print()

# Accessing data in DataFrame
print("Accessing data in DataFrame:")
print(df['Name'])  # Accessing a column
print(df.iloc[0])  # Accessing a row using integer-location based indexing
print()

# Example Series
ages = pd.Series([25, 35, 30, 28], name='Age')
print("Series:")
print(ages)
print()

# Accessing data in Series
print("Accessing data in Series:")
print(ages[0])  # Accessing data using index
```

## Q-6: Which of the following is mutable in nature Series, DataFrame, Panel?

# Mutable Structures in Pandas

## 1. Series: 

```python
import pandas as pd

# Creating a Series
s = pd.Series([1, 2, 3, 4])
print(s)

# Modifying a value
s[0] = 10
print(s)
```

## 2) DataFrame:

```python
# Creating a DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
print(df)
```

```python
# Modifying a value
df.loc[0, 'A'] = 10
print(df)
```

```python
# Adding a new column
df['C'] = [7, 8, 9]
print(df)
```

```python
# Dropping a column
df.drop('B', axis=1, inplace=True)
print(df)
```

# Deprecated Structure

## 3) Panel:

```python
# Note: This is just for historical reference. Panel is deprecated.
import pandas as pd
import numpy as np

data = np.random.rand(2, 3, 4)
panel = pd.Panel(data)

# Modifying a value (Note: This will not work in newer versions of pandas)
panel.iloc[0, 0, 0] = 10
print(panel)
```

```python
import pandas as pd

# Create Series for each column
names = pd.Series(['Alice', 'Bob', 'Charlie', 'David'])
ages = pd.Series([25, 30, 35, 40])
cities = pd.Series(['New York', 'Los Angeles', 'Chicago', 'Houston'])

# Create a DataFrame using the Series
df = pd.DataFrame({
    'Name': names,
    'Age': ages,
    'City': cities
})

# Display the DataFrame
print(df)
```
